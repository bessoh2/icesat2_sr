## Check photon count (Cesar's Method)
See manuscript in review for Cesar's methods, using ATL06. I am replicating his efforts using ATL06SR data to check his result's reproducibility.  
  
**Thresholding:** Cesar used 180 and 50 photons for snow on/off cutoffs, for strong and weak beams respectively. Using ATL06SR data cut down on the total number of photons because it only uses ground photons and only uses high confidence photons instead of medium and high photons. So my estimated, proposed thresholds (determined from just looking at the boxplots) are 90 and 22 - about half of Cesar's thresholds.

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import datetime
import pickle
```

```python
atl06sr_gdf = pickle.load(open('../../data/tuolumne_all_sr-atl06-09282022_processed.pkl', 'rb'))
```

```python
strong = [1,3,5]
weak = [2,4,6]

atl06sr_strong = atl06sr_gdf.loc[atl06sr_gdf['spot'].isin(strong)]
atl06sr_weak = atl06sr_gdf.loc[atl06sr_gdf['spot'].isin(weak)]
```

```python
atl06sr_strong['month'] = atl06sr_strong.index.month
```

```python
atl06sr_weak['month'] = atl06sr_weak.index.month
```

```python
atl06sr_strong['n_fit_photons'].hist()
```

```python
atl06sr_strong['n_fit_photons'].describe()
```

```python
atl06sr_weak['n_fit_photons'].describe()
```

```python
fig, ax = plt.subplots()
atl06sr_gdf.boxplot(column=['n_fit_photons'], by=['spot'], showfliers=False, ax=ax)
fig.suptitle('')
#ax.axhline(y=180)
ax.set_title('Tuolumne Strong Beams = 1,3,5')
ax.set_ylabel('n_fit_photons')
```

```python
fig, ax = plt.subplots()
atl06sr_strong.boxplot(column=['n_fit_photons'], by=['month'], showfliers=False, ax=ax)
fig.suptitle('')
ax.axhline(y=180, c='k', linewidth=1)
ax.axhline(y=90, linestyle='--', c='k', linewidth=1)
ax.set_title('Tuolumne Strong Beams')
ax.set_ylabel('n_fit_photons')
```

```python
fig, ax = plt.subplots()
atl06sr_weak.boxplot(column=['n_fit_photons'], by=['month'], showfliers=False, ax=ax)
fig.suptitle('')
ax.axhline(y=50, c='k', linewidth=1)
ax.axhline(y=22, linestyle='--', c='k', linewidth=1)
ax.set_title('Tuolumne Weak Beams')
ax.set_ylabel('n_fit_photons')
```

```python
#Open, Plot, and Explore Raster Data with Python 
#https://www.earthdatascience.org/courses/use-data-open-source-python/intro-raster-data-python/fundamentals-raster-data/open-lidar-raster-python/
#Applied to ASO Snow Off data from Tuolumne Meadows (downloaded from Linux Box Storage)

# Import necessary packages
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Use geopandas for vector data and rasterio for raster data
import geopandas as gpd
import rasterio as rio
# Plotting extent is used to plot raster & vector data together
from rasterio.plot import plotting_extent

import earthpy as et
import earthpy.plot as ep

# Prettier plotting with seaborn
sns.set(font_scale=1.5, style="white")
```

```python
dem_land = os.path.join("ASO",
                        "snow_off",
                        "ASO_snowoff_linux",
                       "mcc_dem_3p0m_agg_TUOtrimmed_MANUAL.tif")
```

```python
ATL06_path = os.path.join("ATL06_simplified/")
```

```python
# Proceess and Open Snow-Off Data
# Open the file using a context manager ("with rio.open" statement)
with rio.open(dem_land) as dem_lnd:
    dtm_lnd_pre_arr = dem_lnd.read(1)
    dtm_lnd_pre_meta = dem_lnd.profile
    dtm_lnd_pre_plot_ext = plotting_extent(dem_lnd)
    print(dem_lnd.crs)
#Clean raster by removing zero values
dtm_lnd_pre_arr[dtm_lnd_pre_arr == (-9999.)] = np.nan
print(dtm_lnd_pre_plot_ext)
```

```python
df1l = pd.read_csv(ATL06_path+'pass_20190813_gt1l.csv', index_col=0)
gdf1l = gpd.GeoDataFrame(df1l,geometry=gpd.points_from_xy(df1l.x, df1l.y))

df1r = pd.read_csv(ATL06_path+'pass_20190813_gt1r.csv', index_col=0)
gdf1r = gpd.GeoDataFrame(df1r,geometry=gpd.points_from_xy(df1r.x, df1r.y))

df2l = pd.read_csv(ATL06_path+'pass_20190813_gt2l.csv', index_col=0)
gdf2l = gpd.GeoDataFrame(df2l,geometry=gpd.points_from_xy(df2l.x, df2l.y))

df3l = pd.read_csv(ATL06_path+'pass_20190813_gt3l.csv', index_col=0)
gdf3l = gpd.GeoDataFrame(df3l,geometry=gpd.points_from_xy(df3l.x, df3l.y))


fig, ax = plt.subplots(figsize=(8, 20))
ep.plot_bands(dtm_lnd_pre_arr, ax=ax, title="Lidar DEM Tuolumne ASO Snow-On \n with 12July ICESAT-2 Track", cmap="Greys", extent=dtm_lnd_pre_plot_ext)  
gdf1l.plot(ax=ax, color='green', markersize=0.01)
gdf1r.plot(ax=ax, color='teal', markersize=0.01)
gdf2l.plot(ax=ax, color='red', markersize=0.01)
gdf3l.plot(ax=ax, color='blue', markersize=0.01)
plt.show()
#plt.savefig('fig.pdf')
```
